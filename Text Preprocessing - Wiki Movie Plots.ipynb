# 1. Basic feature extraction using text data

```python
# read the data
data=pd.read_csv('/kaggle/input/wiki_movie_plots_deduped.csv')
```

```python
data.head()
```

```python
#count the number of words in each plot
data['plot_word_count'] = data['Plot'].apply(lambda x: len(str(x).split(" ")))
data[['Plot','plot_word_count']].head()
```

```python
#count the number of words in each Genre
data['Genre_word_count'] = data['Genre'].apply(lambda x: len(str(x).split(" ")))
data[['Genre','Genre_word_count']].head()
#data[data['Genre_word_count']>1]
```

```python
# No of charecters in plot
data['plot_char_count'] = data['Plot'].str.len() ## this also includes spaces
data[['Plot','plot_char_count']].head()
```

```python
# Average word length
def avg_word(sentence):
  words = sentence.split()
  return (sum(len(word) for word in words)/len(words))

data['plot_avg_word'] = data['Plot'].apply(lambda x: avg_word(x))
data[['Plot','plot_avg_word']].head()
```

```python
# Number of stopwords
from nltk.corpus import stopwords
stop = stopwords.words('english')

data['plot_stopwords_count'] = data['Plot'].apply(lambda x: len([x for x in x.split() if x in stop]))
data[['Plot','plot_stopwords_count']].head()
```

```python
# Number of stopwords :p
from nltk.corpus import stopwords
stop = stopwords.words('english')

data['Genre_stopwords_count'] = data['Genre'].apply(lambda x: len([x for x in x.split() if x in stop]))
data[['Genre','Genre_stopwords_count']].head()
```

```python
import re 

def special_char(string): 
    regex = re.compile('[@_!#$%^&*()<>?/\|}{~:]')    
    if(regex.search(string) != None): 
        return string 

data['spl_char'] = data['Plot'].apply(lambda x: special_char(x))
data[['Plot','spl_char']]
```

```python
data['spl_char_genre'] = data['Genre'].apply(lambda x: special_char(x))
data[['Genre','spl_char_genre']]
```

```python
# Number of numerics
data['numeric_plot_count'] = data['Plot'].apply(lambda x: len([x for x in x.split() if x.isdigit()]))
data[['Plot','numeric_plot_count']].head()
```

```python
# Number of numerics
data['numeric_genre_count'] = data['Genre'].apply(lambda x: len([x for x in x.split() if x.isdigit()]))
data[['Genre','numeric_genre_count']].head()
```

```python
# Number of Uppercase words
data['upper_plot'] = data['Plot'].apply(lambda x: len([x for x in x.split() if x.isupper()]))
data[['Plot','upper_plot']].head()
```

```python
#uppercase words in Genre
data['upper_genre'] = data['Genre'].apply(lambda x: len([x for x in x.split() if x.isupper()]))
data[['Genre','upper_genre']].head()
```

# Basic Pre-processing

```python
# Lower case
data['Plots_Cleaned_A'] = data['Plot'].apply(lambda x: " ".join(x.lower() for x in x.split()))
data[['Plot','Plots_Cleaned_A']].head()
```

```python
#Remove punctuations
data['Plots_Cleaned_A'] = data['Plots_Cleaned_A'].str.replace('[^\w\s]','')
data[['Plot','Plots_Cleaned_A']].head()
```

```python
# Removal of Stop Words
from nltk.corpus import stopwords
stop = stopwords.words('english')
data['Plots_Cleaned_A'] = data['Plots_Cleaned_A'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
data[['Plot','Plots_Cleaned_A']].head()
```

```python
#Common word removal - just looking at 10 most frequently occuring words
freq = pd.Series(' '.join(data['Plots_Cleaned_A']).split()).value_counts()[:10]
freq
```

```python
# remove them if you want to !
freq = list(freq.index)
data['Plots_Cleaned_A'] = data['Plots_Cleaned_A'].apply(lambda x: " ".join(x for x in x.split() if x not in freq))
data['Plots_Cleaned_A'].head()
```

```python
# Rare words removal - they wont have much associations with other words
freq = pd.Series(' '.join(data['Plots_Cleaned_A']).split()).value_counts()[-10:]
freq
```

```python
freq = list(freq.index)
```

```python
data['Plots_Cleaned_A'] = data['Plots_Cleaned_A'].apply(lambda x: " ".join(x for x in x.split() if x not in freq))
data['Plots_Cleaned_A'].head()
```

```python
# Spelling correction
# care should be taken for words like ur(your) etc otherwise text blob will convert it to something else
def clean_text(text):
    text = text.lower()
    text = re.sub(r"what's", "what is ", text)
    text = re.sub(r"\'s", " ", text)
    text = re.sub(r"\'ve", " have ", text)
    text = re.sub(r"can't", "can not ", text)
    text = re.sub(r"n't", " not ", text)
    text = re.sub(r"i'm", "i am ", text)
    text = re.sub(r"\'re", " are ", text)
    text = re.sub(r"\'d", " would ", text)
    text = re.sub(r"\'ll", " will ", text)
    text = re.sub(r"\'scuse", " excuse ", text)
    text = text.strip(' ')
    return text

data['Plots_Cleaned_A'] = data['Plots_Cleaned_A'].apply(clean_text)


```

```python
from textblob import TextBlob
data['Plots_Cleaned_A'][:5].apply(lambda x: str(TextBlob(x).correct()))
```

```python
#  Tokenization
#TextBlob(data['Plots_Cleaned_A']).words
data['Plots_Cleaned_A'][:5].apply(lambda x: str(TextBlob(x).words))
```

```python
# Stemming
from nltk.stem import PorterStemmer
st = PorterStemmer()
data['Plots_Cleaned_A'][5:10].apply(lambda x: " ".join([st.stem(word) for word in x.split()]))
```

```python
#  Lemmatization
from textblob import Word
data['Plots_Cleaned_A'][5:10].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))
```

# Advance Text Processing

```python
# n- grams 
#TextBlob(data['Plots_Cleaned_A'][:1]).ngrams(2) # 2 represents bigrams
data['Plots_Cleaned_A'][:5].apply(lambda x: str(TextBlob(x).ngrams(2)))
```

```python
# Term frequency
# TF = (Number of times term T appears in the particular row) / (number of terms in that row)
tf1 = (data['Plots_Cleaned_A'][1:2]).apply(lambda x: pd.value_counts(x.split(" "))).sum(axis = 0).reset_index()
tf1.columns = ['words','tf']
tf1
```

```python
#IDF - Inverse Document Frequency
# IDF = log(N/n), where, N is the total number of rows and n is the number of rows in which the word was present.
for i,word in enumerate(tf1['words']):
  tf1.loc[i, 'idf'] = np.log(data.shape[0]/(len(data[data['Plots_Cleaned_A'].str.contains(word)])))
tf1
# The more the value of IDF, the more unique is the word.
```

```python
# Term Frequency – Inverse Document Frequency (TF-IDF)
#TF-IDF is the multiplication of the TF and IDF which we calculated above.
tf1['tfidf'] = tf1['tf'] * tf1['idf']
tf1
```
